## **Inicio: cargar datos**

### *Abrir un archivo (csv)*

R permite abrir diferentes tipos de archivos, en este caso, vamos a cargar uno de los tipos más utilizados.

```r
# Abrir archivos CSV

bd <- read.csv(file = "buses.csv", header = TRUE, sep = ";")
head(bd)

# Descripción de los datos
str(bd)
```

Algunos tipos de datos en R son los siguientes:

*   int = números enteros
*   dbl = números decimales
*   chr = caracteres o letras o texto  

El capítulo 4 muestra el resto de tipos de datos.

```r
# Transformación de una columna

bd$COOMICRO.LTDA[is.na(bd$COOMICRO.LTDA)] <- 0

# Transformación de todos los valores

bd[is.na(bd)] <- 0

head(bd)
```

## **Estadísticas descriptivas**

```r
# Promedio
mean(bd$COOMICRO.LTDA)

# Mediana
median(bd$COOMICRO.LTDA)

# Desviación típica

sd(bd$COOMICRO.LTDA)

# Varianza

var(bd$COOMICRO.LTDA)

# Percentiles

quantile(bd$COOMICRO.LTDA)

quantile(bd$COOMICRO.LTDA, 0.50)
```

```r
summary(bd)
```

### *Suma de buses (anual)*

```r
# Suma de todos los buses

bd$total <- bd$COOMICRO.LTDA + bd$COOPTRANSORIENTE + bd$COOTRANSFRONORTE.LTDA + bd$CORTA.DISTANCIA.S.A. + bd$GUASIMALES.S.A + bd$ONTIVEROS.SAS+
  bd$PETROLEA.S.A + bd$RISARALDA.S.A + bd$TONCHALA.S.A + bd$TRANS.ORIENTAL.S.A. + bd$TRASAN.PLUS.S.A.S + bd$TRASAN.S.A.S

sum(bd$total)

head(bd)
```

```r
# Convertir la columna 'periodo' en una variable tipo número
bd$periodo <- as.numeric(as.character(bd$periodo))

summary(bd$total)
```
